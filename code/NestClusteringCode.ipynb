```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from mpl_toolkits.mplot3d import Axes3D

# Define the selected columns for clustering
selected_columns = ['Sex', 'Funder Type', 'Study Type', 'Start Month', 'Start Quarter',
       'Conditions_Category', 'Num_Collaborators', 'Collaborator_Type',
       'Condition_Category_old', 'Sponsor Type', 'CHILD', 'ADULT',
       'OLDER_ADULT', 'Masking', 'Observational Model', 'Time Perspective',
       'Masking Details', 'Allocation_', 'Allocation_NA',
       'Allocation_NON_RANDOMIZED', 'Allocation_RANDOMIZED',
       'Allocation_Unknown', 'Intervention Model_', 'Intervention Model_CROSSOVER',
       'Intervention Model_FACTORIAL', 'Intervention Model_PARALLEL',
       'Intervention Model_SEQUENTIAL', 'Intervention Model_SINGLE_GROUP',
       'Intervention Model_Unknown', 'Masking Level_DOUBLE', 'Masking Level_NONE',
       'Masking Level_QUADRUPLE', 'Masking Level_SINGLE', 'Masking Level_TRIPLE',
       'Primary Purpose_', 'Primary Purpose_BASIC_SCIENCE', 'Primary Purpose_DEVICE_FEASIBILITY',
       'Primary Purpose_DIAGNOSTIC', 'Primary Purpose_ECT', 'Primary Purpose_HEALTH_SERVICES_RESEARCH',
       'Primary Purpose_OTHER', 'Primary Purpose_PREVENTION', 'Primary Purpose_SCREENING',
       'Primary Purpose_SUPPORTIVE_CARE', 'Primary Purpose_TREATMENT', 'Primary Purpose_Unknown',
       'Country', 'Development Category', 'Study Title_Sentiment_Score',
       'Brief Summary_Sentiment_Score']

# Filter the dataset to include only the selected columns
usecase_3_selected = usecase_3[selected_columns].copy()

# Identify categorical and numerical columns
categorical_cols = usecase_3_selected.select_dtypes(include=['object', 'category']).columns
numerical_cols = usecase_3_selected.select_dtypes(include=['int64', 'float64', 'bool']).columns

# Define preprocessing pipeline
preprocessor = ColumnTransformer([
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='median')),  # Fill missing numerical values
        ('scaler', StandardScaler())  # Scale numerical data
    ]), numerical_cols),

    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),  # Fill missing categorical values
        ('encoder', OneHotEncoder(handle_unknown='ignore'))  # Encode categorical variables
    ]), categorical_cols)
])

# Apply preprocessing
X_preprocessed = preprocessor.fit_transform(usecase_3_selected)

# Apply K-Means clustering
optimal_clusters = 2  # Adjust based on elbow method
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_preprocessed)

# Assign cluster labels to the dataset
usecase_3_selected['Cluster'] = clusters
usecase_3['Cluster'] = clusters  # Assign to original dataset

### Perform PCA to reduce dimensions to 2D ###
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_preprocessed)

# Create a DataFrame for plotting
plot_df_2d = pd.DataFrame(X_pca_2d, columns=['PC1', 'PC2'])

# Ensure index alignment
plot_df_2d = plot_df_2d.reset_index(drop=True)
usecase_3 = usecase_3.reset_index(drop=True)

# Assign Cluster and Study Type without index mismatch
plot_df_2d['Cluster'] = clusters
plot_df_2d['Study Type'] = usecase_3['Study Type'].values  # Ensuring index alignment

# 2D Scatter Plot (Color by Study Type, Shape by Cluster)
plt.figure(figsize=(10, 7))
scatter = sns.scatterplot(
    x='PC1', y='PC2', hue=plot_df_2d['Study Type'], style=plot_df_2d['Cluster'],
    palette="tab10", data=plot_df_2d, alpha=0.7, s=100  # Increase marker size
)

plt.title('2D PCA Visualization of Clusters with Study Type Colors', fontsize=19, fontweight="bold")
plt.xlabel('Principal Component 1', fontsize=19, fontweight="bold")
plt.ylabel('Principal Component 2', fontsize=19, fontweight="bold")

# Customize legend
legend = plt.legend(loc='upper right', fontsize=16, frameon=True, edgecolor='black')

# Increase legend marker sizes
#for handle in legend.legendHandles:
#    handle.set_markersize(15)  # Adjust icon size in legend

plt.show()
```

```python
import pandas as pd
import numpy as np
import joblib
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

# Define the selected columns for clustering
selected_columns = ['Sex', 'Funder Type',
       'Study Type', 'Start Month', 'Start Quarter',
       'Conditions_Category', 'Num_Collaborators', 'Collaborator_Type',
       'Condition_Category_old', 'Sponsor Type', 'CHILD', 'ADULT',
       'OLDER_ADULT', 'Masking', 'Observational Model', 'Time Perspective',
       'Masking Details', 'Allocation_', 'Allocation_NA',
       'Allocation_NON_RANDOMIZED', 'Allocation_RANDOMIZED',
       'Allocation_Unknown', 'Intervention Model_',
       'Intervention Model_CROSSOVER', 'Intervention Model_FACTORIAL',
       'Intervention Model_PARALLEL', 'Intervention Model_SEQUENTIAL',
       'Intervention Model_SINGLE_GROUP', 'Intervention Model_Unknown',
       'Masking Level_DOUBLE', 'Masking Level_NONE', 'Masking Level_QUADRUPLE',
       'Masking Level_SINGLE', 'Masking Level_TRIPLE', 'Primary Purpose_',
       'Primary Purpose_BASIC_SCIENCE', 'Primary Purpose_DEVICE_FEASIBILITY',
       'Primary Purpose_DIAGNOSTIC', 'Primary Purpose_ECT',
       'Primary Purpose_HEALTH_SERVICES_RESEARCH', 'Primary Purpose_OTHER',
       'Primary Purpose_PREVENTION', 'Primary Purpose_SCREENING',
       'Primary Purpose_SUPPORTIVE_CARE', 'Primary Purpose_TREATMENT',
       'Primary Purpose_Unknown', 'Country', 'Development Category',
       'Study Title_Sentiment_Score', 'Brief Summary_Sentiment_Score'
       ]

# Filter the dataset to include only the selected columns
usecase_3_selected = usecase_3[selected_columns].copy()

# Identify categorical and numerical columns
categorical_cols = usecase_3_selected.select_dtypes(include=['object', 'category']).columns
numerical_cols = usecase_3_selected.select_dtypes(include=['int64', 'float64', 'bool']).columns

# Define preprocessing pipeline
preprocessor = ColumnTransformer([
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='median')),  # Fill missing numerical values
        ('scaler', StandardScaler())  # Scale numerical data
    ]), numerical_cols),

    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),  # Fill missing categorical values
        ('encoder', OneHotEncoder(handle_unknown='ignore'))  # Encode categorical variables
    ]), categorical_cols)
])

# Define the full pipeline including K-Means clustering
optimal_clusters = 3  # Adjust based on elbow method
model_pipeline = Pipeline([
    ('preprocessor', preprocessor),  # Apply preprocessing
    ('kmeans', KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10))  # K-Means Clustering
])

# Train the model
model_pipeline.fit(usecase_3_selected)

# Save the trained pipeline to a file
joblib.dump(model_pipeline, "clustering_model.pkl")
print("Model saved as 'clustering_model.pkl'")
```
